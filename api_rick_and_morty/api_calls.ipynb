```python
def gender_prediction(name):
    base_url = 'https://api.genderize.io?name='
    full_url = f"{base_url}{name}"
    response5 = requests.get(full_url)
    if response5.status_code == 200:
        data = response5.json()
        name = data['name']
        gender = data['gender']
        probability = data['probability']
        count = data['count']
        msg = print(f"""Hello {name},

        We believe you are a {gender} cos we know {count} other people named {name} and they are all {gender}s.

        Thank You!""")
    return msg
```
